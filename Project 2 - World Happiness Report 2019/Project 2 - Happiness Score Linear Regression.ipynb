```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

Read the csv file to a dataframe.

```python
df = pd.read_csv('happy2019.csv')
```

```python
df.head(3)
```

```python
df.columns.values
```

Prepare the data to be tested. Drop columns not required. Separate x-variables and the y-variable.

```python
Drop = ['Overall rank', 'Country', 'Region', 'Latitude', 'Longitude']
dfEvaluate = df.drop(Drop,axis=1)
X = dfEvaluate.drop('Score',axis=1)    # Predictors/Input Variables/DV
y = dfEvaluate['Score']    # Target/Label/IV  
```

Split the data into training and testing data set. This model tests with 60% training and 40% testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=101)  
```

Initiate model and feed training data

```python
bhr_model = LinearRegression() 
bhr_model.fit(X_train,y_train)  
```

```python
test_predictions = bhr_model.predict(X_test)
```

Series of variable coefficients.

```python
df_bhr_coeff = pd.DataFrame(bhr_model.coef_,X.columns,columns=['Coefficient'])
df_bhr_coeff
```

Create a list of evaluation metrics for the linear regression. 

```python
MAE = mean_absolute_error(y_test,test_predictions)
MSE = mean_squared_error(y_test,test_predictions)
RMSE = np.sqrt(MSE)
R_SQR = r2_score(y_test, test_predictions)
yINT = bhr_model.intercept_
printDict = {MAE: 'Mean Absolute Error: ', MSE:'Mean Squared Error: ',
             RMSE:'Square Root Mean Squared Error: ', R_SQR: 'R^2: ', 
             yINT:'Y-Intercept: '}
printList = [MAE,MSE,RMSE,R_SQR,yINT]
```

Evaluation: This model is a good fit with a r^2 of 61%. In the domain of human analytics, 61% is a high score. 

```python
for item in printList:
    print(printDict.get(item), round(item, 2))
```
